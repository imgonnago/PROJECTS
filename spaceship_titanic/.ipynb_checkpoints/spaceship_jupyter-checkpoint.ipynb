```python
#원 핫 인코딩 대신 라벨인코딩 사용. 이유는 원핫인코딩을 하면 히트맵을 볼 때 컬럼이 너무 많아져서 보기 복잡하고 컬럼 자체의 상관관계를 내타내기 힘듦
#라벨인코더는 숫자로 표현되서 컬럼 자체의 의미는 조금 퇴색되어도 상관관계를 높일 수 있음
```

```python
df_train_label = df_train_drop.copy()
df_test_label = df_test_drop.copy()
scaler = LabelEncoder()

for col in df_train_label.select_dtypes(include='object'):
    df_train_label[col] = scaler.fit_transform(df_train_label[col])
    
for col in df_test_label.select_dtypes(include='object'):
    df_test_label[col] = scaler.fit_transform(df_test_label[col])
    
print(df_train_label)
print(df_test_label)
```

```python
df_train_label = df_train_label.fillna(df_train_label.mode().iloc[0])
df_test_lable = df_test_label.fillna(df_test_label.mode().iloc[0])
```

```python
#남은 수치형 컬럼에서 이상치가 존재하는 확인 후 스케일링 진행 여부 확인
```

```python
#스케일링 전 train, test셋 split
```

```python
x = df_train_label[['CryoSleep', 'RoomService', 'Spa', 'VRDeck']]
y = df_train_label['Transported']
train_x, test_x, train_y,test_y = train_test_split(x,y, test_size=0.3, random_state=42)
train_x.shape, train_y.shape,test_x.shape, test_y.shape
```

```python
#이상치가 많이 보이므로 스케일링 진행. minmax로 0,1사이의 값으로 스케일링
```

```python
minmax = MinMaxScaler()
train_x = minmax.fit_transform(train_x)
test_x = minmax.transform(test_x)
print(train_x)
print(test_x)
```

```python
#모델 정의 
#rf ===== acc =====
#0.7638036809815951
```

```python
automl = AutoML()

settings = {
    "time_budget": 300,
    "metric": 'accuracy',
    "task":'classification',
    "seed":42
}

automl.fit(X_train=train_x, y_train=train_y, **settings)
print(f"최적의 모델: {automl.best_estimator}")
```

```python
rf = RandomForestClassifier(
    n_estimators=1000,
    max_depth=50,
    min_samples_leaf=3,
    random_state=42
)

rf.fit(train_x, train_y)
```

```python
rf_pred = automl.predict(test_x)
rf_acc = accuracy_score(test_y, rf_pred)
rf_mat = confusion_matrix(test_y, rf_pred)
rf_rep = classification_report(test_y, rf_pred)
print(f'===== acc =====')
print(rf_acc)
print(f'=====matrix=====')
print(rf_mat)
print(f'=====report=====')
print(rf_rep)
```
